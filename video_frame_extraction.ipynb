```python
violence=pd.DataFrame()

for i, file in enumerate(voilence_file):
    violence=violence.append({'video_name':file}, ignore_index=True)
    
    
```

```python
for i in tqdm(range(violence.shape[0])):
    count=0
    videoFile=violence['video_name'][i]
    cap=cv2.VideoCapture(voilencePath+videoFile)
    frameRate=cap.get(5)
    x=1
    while(cap.isOpened()):
        frameId=cap.get(1)
        ret, frame=cap.read()
        if(ret!=True):
            break
        if(frameId%math.floor(frameRate)==0):
            filename=violence_images+videoFile+'_frame%d.jpg' %count;count+=1
            cv2.imwrite(filename, frame)
    cap.release()
```
